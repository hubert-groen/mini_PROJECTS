```python
import pandas as pd

df_original = pd.read_csv('FINANSES_original.csv')
df_original.columns
```

```python
# DROPPING UNNECESARRY COLUMNS

important_columns = ['Country', 'Partner', 'Value']

df_important_columns = df_original[important_columns].copy()

df_important_columns['Country'] = df_important_columns['Country'].str.lower()
df_important_columns['Partner'] = df_important_columns['Partner'].str.lower()

df_important_columns
```

```python
# ABSOLUTE VALUE

df_important_columns['Value'] = df_important_columns['Value'].abs()

df_important_columns
```

```python
# ONE-DIRECTION FLOW 

def one_direction_flow(df):

    indexes_to_drop = []
    counter = 0
    el = 0

    for index, row in df.iterrows():

        if index in indexes_to_drop:
            continue

        country = row['Country']
        partner = row['Partner']

        repeat_index = df[(df['Country'] == partner) & (df['Partner'] == country)].index

        
        if not repeat_index.empty:

            counter += 1
            print('------------')
            print(country)
            print(partner)
            value = df.loc[repeat_index, 'Value'].values[0]
            print(value)
            print('------------')

            df.at[index, 'Value'] += value

            indexes_to_drop.extend(repeat_index)
        
        else:
            el += 1

    print(counter)
    print(el)
    print(len(indexes_to_drop))
    df = df.drop(indexes_to_drop)

    return df


df_one_direction = one_direction_flow(df_important_columns)

df_one_direction['Value'] = df_one_direction['Value'].round(1)
```
